dy/dx(f(x)) = f(x[i]) - f(x[i-1])/ x[i] - x[i-1]
dt/dx(f(x)) = f(x[i+1]) - f(x[i])/ x[i+1] -x[i]
parâmetros:
- função
- numero de pontos
- intervalo
- lista de pontos de x
- lista de pontos de f(x)
- lista de pontos de dy/dx(f(x))

Plotar gráfico de dy/dx(f(x)) para f(x) = sin(x)

```python
from numpy import arange, linspace, array
from matplotlib import pyplot as plt
from math import sin, cos, pi
```

```python
f = lambda x: sin(x)
n = 101
xi = -1*pi
xf = 1*pi
x = [i for i in linspace(xi, xf, n)]
dx = (xf-xi)/ n-1
```

```python
fx = []
for i in x:
    fx.append(f(i))
```

```python
dydx = []
for i in range(0, n-1):
    dydx.append((f(x[i+1]) - f(x[i]))/ x[i+1] - x[i])
```

```python
x.remove(x[len(x)-1])
fx.remove(fx[len(fx)-1])
```

```python
print(len(x))
print(len(fx))
print(len(dydx))
plt.plot(x, dydx)
print(x, end='\n\n')
print(dydx)
```

```python
fig, ax = plt.subplots()
ax.plot(x, fx, label='f(x)')
ax.plot(x, dydx, label='dy/dx')
ax.legend()
plt.show()
```
