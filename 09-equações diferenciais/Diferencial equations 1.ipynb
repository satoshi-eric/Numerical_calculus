Faça um gráfico de x(t) e v(t) para o seguinte caso:
        * m = 1
        * k = 1
        * x0 = 1
        * v0 = 0
        * tf = 10
        * n = 101
        * deltaT = 0.1
        compare com a solução exata

```python
# importação de bibliotecas
from math import cos
from matplotlib import pyplot as plt
from numpy import arange, linspace
```

```python
#definição de variáveis
m = 1 # massa
k = 1 # constatnte elástica
x0 = 1 # posição 0 (inicial) [será parte de uma lista]
v0 = 0 # velicidade 0 (inicial) [será parte de uma lista]
tf = 10 # tempo final
ti = 0 # tempo inicial
n = 101 #número de pontos
delta_t = 0.1 # variação de tempo
```

```python
# definição de funções
v = lambda x, v: v - (k/m) * x * delta_t
x = lambda x, v: x + v * delta_t
```

```python
# declaração das listas para plotar os gráficos
vt = [v0]
xt = [x0]
# criação das listas a partir das funções definidas e dos pontos v0 e x0
for i in arange(0, 100, 1):
    vt.append(v(xt[i], vt[i]))
    xt.append(x(xt[i], vt[i]))
```

```python
# definição de lista para o tempo
t = [i for i in linspace(ti, tf, n)]
```

```python
# lista para a resposta
g = lambda x: cos(x)
res = [g(i) for i in linspace(ti, tf, n)]
```

```python
fig, ax = plt.subplots()
ax.plot(t, xt, label = "posição")
ax.plot(t, vt, label = "velocidade")
ax.plot(t, res, label = "resposta")
ax.legend()
plt.show()
```
