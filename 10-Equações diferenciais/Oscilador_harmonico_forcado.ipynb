```python
from matplotlib import pyplot as plt
from numpy import zeros, linspace, arange
from math import cos
```

```python
dt = 0.01
m = 1.0
w0 = 0.5
A = 0.1
x0 = 1
v0 = 0
we = 0.5
ti = 0
tf = 100
n = int(((tf - ti)/dt)+1)
```

```python
f = lambda i: A*cos(we*dt*i)
v = lambda v, x, f: v + ((f/m)-(w0**2)*x)*dt
x = lambda v, x: x + dt*v
```

```python
ft = [f(i) for i in range(n)]
vt = zeros(n)
xt = zeros(n)
vt[0] = v0
xt[0] = x0
for i in arange(1, n, 1):
    vt[i] = v(vt[i-1], xt[i-1], ft[i-1])
    xt[i] = x(vt[i-1], xt[i-1])
```

```python
t = [i for i in linspace(ti, tf, n)]
```

```python
fig, ax = plt.subplots()
ax.plot(t, xt, label='x(t)')
ax.legend()
plt.show()
```
