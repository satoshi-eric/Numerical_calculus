```python
# importação de bibliotecas
from math import e
from numpy import arange
from matplotlib.pyplot import plot, show, scatter
```

```python
# definindo limites para a função
# o limite pode ser pequeno, mas também pode ser grande, depende da funçao
lim_min = -2.5
lim_max = 2.5
# definição da função 
f = lambda x: e**-x**2
# criando uma lista de numeros para x de lim_min até lim_max
x = [i for i in arange(lim_min, lim_max, 0.1)]
# mapeando uma lista de numeros correspondentes aos das variáveis com a função
y = list(map(f, x))
```

```python
# plotando o gráfico
plot(x, y)
# mostrando o gráfico
show()
```
